## AML 22-23 S2W7 Webinar: Ensembling

### Key Points

* "wisdom of crowds" (too popular-sciency, but let us go with it): if we have models that **perform similarly** but **differently in the data space**, let us wrap them all together into a single model that averages/blends/combines their predictions.

* **averaging/voting** and **stacking** are the simplest strategies to put in place in applied ML.

    - you would train the constituent models in the same way you have been working with your individual models so far:
        + hyper-parameterisation
        + hyper-parameter search
        + train/test split
        + cross validation...

    - heterogeneous: you can assemble models of different types, generated by different algorithms, and so on.
